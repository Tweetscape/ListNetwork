# Create Cryptogurus Network

```python
import plotly.offline as py
import plotly.graph_objects as go
import networkx as nx
```

```python
cryptonet = nx.DiGraph()
```

```python
# Add a node for each user
for user in cryptogurus_dict.keys():
    cryptonet.add_node(user)
```

```python
# For each following, add an edge
# add_edge(a, b) = a follows b

for user_a in cryptogurus_dict.keys():
    for user_b in cryptogurus_dict[user_a]:
        cryptonet.add_edge(user_a, user_b)
```

```python
# Get positions for the nodes in follownet
pos_ = nx.spring_layout(cryptonet)
```

```python
def make_edge(x, y, text):
    
    '''Creates a scatter trace for the edge between x's and y's with given width

    Parameters
    ----------
    x    : a tuple of the endpoints' x-coordinates in the form, tuple([x0, x1, None])
    
    y    : a tuple of the endpoints' y-coordinates in the form, tuple([y0, y1, None])
    
    width: the width of the line

    Returns
    -------
    An edge trace that goes between x0 and x1 with specified width.
    '''
    return  go.Scatter(x         = x,
                       y         = y,
                       line      = dict(width = 1, color = 'dodgerblue'),
                       opacity   = 0.2,
                       hoverinfo = 'text',
                       text      = ([text]),
                       mode      = 'lines')
```

```python
# For each edge, make an edge_trace, append to list
edge_trace = []
for edge in cryptonet.edges():
    user_a = edge[0]
    user_b = edge[1]

    x0, y0 = pos_[user_a]
    x1, y1 = pos_[user_b]

    text   = user_a + '->' + user_b

    trace  = make_edge([x0, x1, None], [y0, y1, None], text)

    edge_trace.append(trace)
```

```python
# Make a node trace
node_trace = go.Scatter(x         = [],
                        y         = [],
                        text      = [],
                        textposition = "top center",
                        textfont_size = 10,
                        mode      = 'markers+text',
                        hoverinfo = 'none',
                        marker    = dict(color = [],
                                         size  = [],
                                         line  = None))
# For each node in follownet, get the position and size and add to the node_trace
for user in cryptonet.nodes():
    x, y = pos_[user]
    node_trace['x'] += tuple([x])
    node_trace['y'] += tuple([y])
    node_trace['marker']['color'] += tuple(['dodgerblue'])
    node_trace['marker']['size'] += tuple([10])
    node_trace['text'] += tuple(['<b>' + user + '</b>'])
```

```python
layout = go.Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)', 
    xaxis =  {'showgrid': False, 'zeroline': False},
    yaxis = {'showgrid': False, 'zeroline': False}, 
)


fig = go.Figure(layout = layout)

for trace in edge_trace:
    fig.add_trace(trace)

fig.add_trace(node_trace)

fig.update_layout(showlegend = False)

fig.update_xaxes(showticklabels = False)

fig.update_yaxes(showticklabels = False)

fig.show()
```
